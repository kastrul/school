```python
from __future__ import print_function

import logging
import warnings

import gensim
import pandas as pd
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.ERROR)

warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
ENCODING = 'iso-8859-1'
COL_Q = ['Id', 'Body']
COL_A = ['ParentId', 'Body']
N_ROWS = 100000
PATH_ANSWERS = 'data/stacksample/Answers.csv'
PATH_QUESTIONS = 'data/stacksample/Questions.csv'
path_TAGS = 'data/stacksample/Tags.csv'

# df_answers = pd.read_csv(PATH_ANSWERS, encoding=ENCODING, low_memory=False, usecols=['Body'], iterator=True)
df_questions = pd.read_csv(PATH_QUESTIONS, encoding=ENCODING, low_memory=False, usecols=COL_Q, nrows=N_ROWS)
df_tags = pd.read_csv(path_TAGS, encoding=ENCODING, low_memory=False)
```

```python
def print_state(current, target):
    print('\rRow {0}/{1}'.format(current, target), end='')
```

```python
def remove_stopwords(sentence):
    return [word for word in sentence if word not in stop_words]


def prepare_sentence(sentence):
    return remove_stopwords(gensim.utils.simple_preprocess(str(sentence), deacc=True))
```

```python
def prepare_data_for_csv(text_data, tag_data):
    dataset = {
        'text': [],
        'tags': []
    }
    n_rows = len(text_data)
    for idx, row in enumerate(text_data):
        print_state(idx + 1, n_rows)
        text = prepare_sentence(row[1])
        tags = tag_data[tag_data['Id'] == row[0]]['Tag'].values
        dataset['text'].append(text)
        dataset['tags'].append(tags)
    return dataset
```

```python
data = prepare_data_for_csv(df_questions.values, df_tags)
```

```python
df = pd.DataFrame(data=data)
```

```python
class MySentences(object):
    def __init__(self, path, columns):
        self.columns = columns
        self.path = path
        self.dictionary = gensim.corpora.Dictionary(
            pd.read_csv(path, encoding=ENCODING, low_memory=False, usecols=columns, chunksize=1000))

    def __iter__(self):
        for chunk in pd.read_csv(self.path, encoding=ENCODING, low_memory=False, usecols=self.columns, chunksize=1):
            yield self.dictionary.doc2bow(tokenize_sentence(chunk))
```
