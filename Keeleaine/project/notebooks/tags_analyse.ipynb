```python
import logging
import pprint
import pandas as pd
import matplotlib.pyplot as plt

logging.basicConfig(format='\r%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

pp = pprint.PrettyPrinter(indent=4)
ENCODING = 'iso-8859-1'
```

```python
PATH_TAGS = '../data/stacksample/Tags.csv'
tags_df = pd.read_csv(PATH_TAGS, encoding=ENCODING, usecols=['Tag'])
```

```python
print(tags_df.mode())
```

```python
tags_df['Tag'].value_counts().nlargest(10).index.values
```

```python
plt.figure(figsize=(15, 10))
tags_df['Tag'].value_counts().nlargest(50).plot(kind='bar')
plt.savefig('../figures/top50_tags.png', dpi=200)
```
